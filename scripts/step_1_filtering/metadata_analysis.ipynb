## Import Libraries and Dir

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Data

```python
METADATA_DIR = '/home/zxia15/data_zxia15/russian-semantics/work/metadata_russian_documents.jsonl.gz'
metadata_df = pd.read_json(METADATA_DIR, lines=True)
metadata_df.head()
```

```python
len(metadata_df)
```

## Stats for Russian-language ratio

```python
text_ratios = metadata_df['russian text ratio']
plt.hist(x=text_ratios, bins=100, density=True)
plt.show()
```

```python
filtered_non_zero_text_ratio = [ratio for ratio in text_ratios if ratio != 0]
plt.hist(x=filtered_non_zero_text_ratio, bins=100, density=True)
plt.show()
```

## Stats for Years (currently accessible ones)

```python
year_stats = metadata_df['year']
filtered_year_stats = [year_stat for year_stat in year_stats if ~np.isnan(year_stat)]
print(f"the most ancient year available is {min(filtered_year_stats)}, and the most recent year is {max(filtered_year_stats)}")
plt.hist(x=filtered_year_stats, bins=100)
plt.show()
```

## Stats for Forms (currently accessible ones)

```python
form_stats = metadata_df['form'].value_counts()
plt.bar(form_stats.index, form_stats.values)
plt.xlabel("Literary Forms")
plt.ylabel("Ratio of the Genre")
for index, value in enumerate(form_stats.values):
    plt.text(index, value, str(value), ha='center', va='bottom', fontsize=6)
plt.xticks(fontsize=7, rotation=45)
plt.yticks(fontsize=7)
plt.show()
```
